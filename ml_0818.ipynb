## 2) 데이터 불러오기 
 
###  sklearn 패키지 내장 데이터 있음




```
from sklearn.datasets import load_iris 

iris=load_iris()   #bunch 형태의 데이터셋으로 구성되어 있음 
print(iris)

iris_data= iris.data  # 독립변수로만 구성된 데이터 
iris_label = iris.target  #종속변수로만 구성된 데이터
print('iris target 값:', iris_label[[0,50,100]])
print('iris target 명:', iris.target_names)  
```

##3) 데이터 전처리 
- 상황:  numpy 형태 , 리스트
- 방법 : pandas 활용해서 데이터 프레임 만들기 
- stackoverflow : load iris dataset using pandas 영어로 구글링 해보고, 코드들 알아보고 How to convert a Scikit-learn dataset to a Pandas dataset? 이런식으로 stack overflow에서 쓸만한 코드 찾기. 이런식으로 공부해야 되고, 나중에 이런 자료를 함수화 시킬수있으면 github에 update해버리면 됨 (다시슬수도있고 포트폴리오도되고)


```
import pandas as pd
import numpy as np
data = load_iris()
df = pd.DataFrame(data['data'], columns=data['feature_names'])
df['target'] = data['target']
df.head()
```

## 4) 데이터 시각화
iris target의 0,1,2가 구체적으로 뭔지 모르니까
- 시각화를 위한 전처리를 할 것 -> 데이터 복사 

```
temp_data =df.copy()
```

```
temp_data.columns=['sepal_length','sepal_width','petal_length','peta_width','species']
```

```
# column명 바꾸기 
temp_data.columns
```

```
temp_data.info
```

```
#map
#0-> "setosa"
temp_data.info
replace_fct={0:"setosa", 1:"versicolor",2:'virginica'}

temp_data['species']=temp_data['species'].map(replace_fct)
print(temp_data.head())
```

```
import matplotlib.pyplot as plt
# 수치형 데이터 , 히스토그램, & 박스플롯 & 산점도 
```

## 5) 데이터셋 분리
# 학습용 데이터와 테스트용 데이터 분리

```
from sklearn.model_selection import train_test_split
```

```
# 훈련데이터와 테스트 데이터 분리 
X_train, X_test, y_train, y_test=train_test_split(iris_data,iris_label,test_size=0.2,random_state=1)
```

- arrays ~
- test_size , 비율 (훈련데이터/ 테스트데이터) 
-random_state는 실험의 재현성을 위한 코드 

## 6) 머신러닝 모형 학습
- 머신러닝 모형을 만든다. 

```
from sklearn.tree import DecisionTreeClassifier

# 객체생성

dt_clf=DecisionTreeClassifier(random_state=11,max_depth=2) 
# max_depth=none 에서 2라고 바꿨는데, 뿌리(root_date) 2개까지 내린걸로 바꿔준거. none이면 끝까지 다 나옴 


#학습수행 

dt_clf.fit(X_train, y_train)
```

## 7) 모형 테스트
#  모형 예측 

```
from sklearn.metrics import accuracy_score
pred=dt_clf.predict(X_test)

print("The accuracy of the Decision Tree is : {:3f}".format(accuracy_score(pred,y_test)))
```

- 시각화 

```
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import plot_confusion_matrix

labels = ['setosa', 'versicolor', 'virginica']
disp = plot_confusion_matrix(dt_clf, 
                             X_test, y_test, 
                             display_labels=labels, 
                             cmap=plt.cm.Reds, 
                             normalize=None)
disp.ax_.set_title('Confusion Matrix');
```

### 모형 결과의 시각화(reporting)





```
import matplotlib.pyplot as plt
explt_vars =["sepal_length","sepal_width","petal_length","petal_width"]
fct_val={0:'setos',1:'versicolor', 2:'virginica'}

plt.figure(figsize=(10,8))
plot_tree(dt_clf, feature_names = explt_vars, class_names = fct_val, filled = True);
```
